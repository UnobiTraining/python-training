# Variables

Variables are named objects that store values.

Examples of values include:

- string: textual content
- integer: whole numbers
- float: decimal numbers
- boolean: `True` or `False`
- `None`: absence of a value

Variables allow programs to store information and reuse it later.

## Variable Naming

Variables should have clear, descriptive names.

Python programmers typically use **lowercase words with underscores**.

Good examples:

- `age = 30`
- `user_name = "Alice"`
- `account_balance = 105.75`

Bad examples:

- `x = 30`
- `thing = "Alice"`
- `data = 105.75`

These work, but they do not explain what the variable contains.

Variables cannot:

- start with a number
- contain spaces
- use reserved Python keywords

````python
age = 30
user_name = "Alice"
account_balance = 105.75

print(age)
print(user_name)
print(account_balance)
````

Examples that will **not** work:

- `2name = "Alice"`
- `user name = "Alice"`
- `class = "Math"`

## Assignment vs Comparison

The equals sign (`=`) assigns a value to a variable.

```python
age = 30
```

This means: store `30` in the variable `age`.

The double equals (`==`) compares values.

```python
age == 30
```

This checks whether `age` is equal to `30`.

````python
age = 30

print(age)
print(age == 30)
print(age == 25)
````

## Strings

A string is textual content.

Strings must be enclosed in quotes.

Python allows:

- single quotes
- double quotes
- triple quotes

````python
name = "Bob"
alt_name = 'Alice'

phrase1 = 'The cashier said "hello" when the customer walked in.'
phrase2 = "Roger's car was parked in the lot across the street."
phrase3 = '''Sarah's dog barked "woof" when Ted threw the ball.'''

print(name)
print(alt_name)
print(phrase1)
print(phrase2)
print(phrase3)
````

## Integers

Integers are whole numeric values.

Numbers are written without quotes.

If quotes are used, Python treats the value as text instead of a number.

````python
number = 5
text_number = "5"

print(number)
print(text_number)

print(type(number))
print(type(text_number))
````

## Python is Dynamically Typed

Python is a dynamically typed language.

A variable is not locked to a single type.
The same variable can hold a string first, and then an integer later.

````python
a = "5"
print(a)
print(type(a))

a = 5
print(a)
print(type(a))
````

## Casting

Sometimes a value must be converted from one type to another.

- `int()` converts to an integer
- `float()` converts to a decimal number
- `str()` converts to text

This is useful when a value looks like a number, but Python is treating it as text.

````python
a = "5"
b = 2

print(type(a))

a = int(a)

print(type(a))
print(a + b)
````

## Variables in Expressions

Variables are often used together in expressions.

The types must make sense together.

A number and a string cannot be added directly.

````python
a = 5
b = "2"

# This would cause an error:
# print(a + b)

b = int(b)
print(a + b)
````

## Boolean Values

A Boolean value is either `True` or `False`.

These are commonly used in comparisons and program logic.

````python
is_active = True
is_admin = False

print(is_active)
print(is_admin)
print(type(is_active))
````

You can also create Boolean values by comparing things.

````python
age = 30

print(age == 30)
print(age == 18)
print(5 == "5")
````

## None

`None` represents the absence of a value.

It is often used as a placeholder when a variable exists but does not have a value yet.

````python
result = None

print(result)
print(type(result))
````

In a conditional statement, `None` evaluates the same way as `False`.

````python
a = None

if a:
    print("This will not run")
else:
    print("This runs because a is None")
````

## Final Example

This example combines strings, integers, and casting.

````python
name = "Alice"
age = "30"

age = int(age)

print(name)
print(age)
print(type(name))
print(type(age))
````
